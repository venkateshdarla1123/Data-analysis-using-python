```python
#create a list of random integers i need 10 elements print their squares save this squares in a new list
l1=[12,9,8,7,4,44,54,21,99,100]
l2=[]
for i in l1:
  l2.append(i*i)
print(l2)
```

```python
#using list comprehension
l1=[12,9,8,7,4,44,54,21,99,100]
l2=[i*i for i in l1]
print(l2)
```

```python
#create a list of random integers print their squares save this squares in a new list if the elements in list is >50
l1=[22,33,44,55,66,77,88]
l2=[]
for i in l1:
  if(i>50):
    l2.append(i * i)
print(l2)
```
