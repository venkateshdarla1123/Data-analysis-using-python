```python
#TUPLES
#Conversion from list to tuple
l=[1,2,3,4]
t=tuple(l)
print(t)
```

```python
#indexing
print(t[3])
```

```python
#comma represents it is a tuple
a=1,
print(type(a))

#length of tuple
print(len(t))

#tuple slicing
print(t[:2])

#allocating values
a,b,c=(1,2,3)
print(a)
print(b)
print(c)

#membership operator in tuple
print(3 in t)
t=list(t)
t.append(9)
t=tuple(t)
print(t)
```

```python
#sets
s={1,2,3,4}
print(s)
```

```python
#no duplicate values in a set
l=["hii","here","hii"]
s2=set(l)
print(s2)
```

```python
#adding elements in a set
s3={1,2,3}
s3.add(4)
print(s3)
#for inserting multiple data
s4={3,6,9}
s4.update([1,2,3])
print(s4)

```

```python
#removing an element
s4.discard(2)
print(s4)
```

```python
#union
a={1,2,3,4,6}
b={1,2,6,9}
print(a.union(b))
#intersection
print(a.intersection(b))
#difference
print(a.difference(b))
#Symmetrics
print(a.symmetric_difference(b))
print(a.issuperset(b))
print(a.issubset(b))
print(a.isdisjoint(b))
```
